```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#Funcion del potencial
PotentialFunction = lambda x,y,r: 1/(np.sqrt((x-r[0])**2 + (y-r[1])**2)) - 1/(np.sqrt((x-r[0])**2 + (y+r[1])**2))


#Funcion de la derivada

def derivative_function(x,y,r,h,function):
    
    #derivada respecto a X:
    diff_x=(function(x+h,y,r) - function(x-h,y,r)) / 2*h
    
    diff_y= (function(x,y+h,r) - function(x,y-h,r)) / 2*h
    
    result=np.asanyarray([-diff_x,-diff_y])
    return result

```

```python
# cracion de los datos
h=0.05
x=np.arange(0.,1.,h)
y=np.arange(0.,1.,h)
r=[0.51,0.21]
# se crea un array donde de la forma [x,y,dx,dy]
datos=np.zeros((len(x)*len(y),4))
# Guardar datos
for i in range(len(x)):
    for j in range(len(y)):
        index = (i * len(y)) + j
        datos[index,0]=x[i]
        datos[index,1]=y[j]
        CampoE=derivative_function(x[i],y[j],r,h,PotentialFunction)
        datos[index,2]=CampoE[0]
        datos[index,3]=CampoE[1]

    
```

```python
#Graficar

for i in range(len(x) * len(y)):
    plt.quiver(datos[i,0],datos[i,1],datos[i,2],datos[i,3],color='red')
    
plt.scatter(r[0],r[1],color='blue')
```
