```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def Funcexacta(t,alfa):
    func= np.e**(t*alfa)
    return func

def funcaproximada(delta,k,alfa,u0):
    
    func = ((1+delta)**k)*u0
    
    return func

#definimos variables
texacto=np.arange(0,10,0.01)
k=np.arange(0,6,1)
alfa=-1
u0=Funcexacta(0,alfa)
#hacemos la grafica
plt.plot(texacto,Funcexacta(texacto,alfa),label='exacta')
plt.plot(k,funcaproximada(1.1,k,alfa,u0),label='dt=1.1')
plt.plot(k,funcaproximada(1.5,k,alfa,u0),label='dt=1.5')
plt.plot(k,funcaproximada(1.9,k,alfa,u0),label='dt=1.9') 
plt.legend()
```
