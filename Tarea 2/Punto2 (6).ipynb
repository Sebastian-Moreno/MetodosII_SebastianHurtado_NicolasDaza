```python
import numpy as np 
import matplotlib.pyplot as plt

def VerletMethod(t,omega,ic):
    h= t[1]-t[0]
    r=np.zeros(len(t))
    dif=np.zeros(len(t))
    r[0]=ic[0]
    r[1]= r[0] + ic[1]*h
    e=(np.cos(np.pi*t) + (np.sin(np.pi*t) )/ np.pi)
    for i in range (2, len(t)):
        dvdt=(omega**2)*(r[i-1])*(-1)
        r[i] = 2 * r[i-1] - r[i-2] + dvdt * (h ** 2)
        dif[i]=np.abs(r[i]-e[i])
    return (r,dif,e)
t = np.linspace(0, 30, 300)
ic=(1,1)
difference=VerletMethod(t,np.pi,ic)[1]
position=VerletMethod(t,np.pi,ic)[0]
exact=VerletMethod(t,np.pi,ic)[2]

plt.plot(t, difference, label="Diferencia entre verlet y solución analítica", color='b')
plt.plot(t, position, label="Verlet Method", color='r')
plt.plot(t, exact, label="Exact Solution", color='g')

plt.xlabel("Tiempo (s)")
plt.ylabel("Distancia (m)")
plt.legend()
plt.grid(True)
plt.show()

"La solución es estable y oscila entre la solución de la exacta"
```
